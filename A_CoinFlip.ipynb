## 동전던지기

coin 을 k 번 던져서 앞, 뒷면이 나오는 결과의 합을 낸다. 이 것을 n 번 반복한다.  

numpy.random.rand 를 이용하고 1(앞면), -1 (뒷면) 으로 정한다. 앞/뒷면 나올 확률이 0.5 라면 k 번의 값을 모두 더한 합계는 0 이 되어야 할 것이다. 정말 그렇게 나오는지 Python 으로 simulation 한다.

1000(k)번 던지기를 100(n) 번 반복한 결과를 histogram 으로 plot 한다. 가로축은 rank 이고 세로축은 빈도를 표시하므로 가로는 합계구간, 세로에는 합계별 값이 출현한 횟수를 표시한다. 이때 1000 번 연속 앞면만 나온 경우 합계가 1000, 1000 번 연속 뒷면만 나온 경우 합계가 -1000 이 될 것이므로 가로축의 구간은 -1000 ~ 1000 으로 한다.

확률이론에 따르면 $|S_k| \geq 4\sqrt{k}$ 일 확률은 $2\times10^{-8}$ 보다 작으므로 histogram 의 bar 들은 $\pm 126.49$ (4 * np.sqrt(1000)) 사이에 위치할 것이다.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_count(k, n):
    """ input: k, generate a sequence of k random coin flips
               n, different trials
        output: return the sum of head(+1) and tail(-1) for n different trials 
        Hint: For simplicity, use 1,-1 to represent head,tails
    """
    x = 2 * (np.random.rand(k, n) > 0.5) - 1
    return x.sum(axis=0)
```

```python
def count_plot(k, n):
    x = generate_count(k,n)
    boundary = 4 * np.sqrt(k)
    plt.hist(x, bins=20, range=(-boundary, boundary));
    plt.plot([-boundary,-boundary], plt.ylim(),'r')
    plt.plot([boundary,boundary], plt.ylim(),'r')
    plt.title('Histogram of coin flip sum when flipping a fair coin {} times'.format(k))
```

```python
count_plot(100, 100)
```

```python
count_plot(1000, 100)
```

```python
count_plot(100000, 100)
```

### 결론

k 개의 random 숫자를 합하는 실험을 하였다. $S_k = x_1 + x_2 + ... + x_k$

$x_i$ 는 1/2 의 확률로 -1 과 +1 이다.

위 실험에 의해 합계 $S_k$ 는 거의 언제나 $\begin{bmatrix}-4\sqrt{k}, +4\sqrt{k}\end{bmatrix}$ 범위에 있음을 확인하였다. 

만약, $k \rightarrow \infty$ 이면 $\frac{4\sqrt{k}}{k} \rightarrow 0$ 이므로, $k \rightarrow \infty$ 이면 $\frac{S_k}{k} \rightarrow 0$ 가 된다.  
즉, k 가 큰 수가 되면 합계는 0 에 수렴하므로 동전의 앞, 뒷면이 나올 확률은 0.5 에 수렴한다.

### 확률이론 (Probability Theory) 는 위와 같은 Monte Carlo simulation 을 거치지 않고, 더 빠르게 정확한 답을 제공한다.
