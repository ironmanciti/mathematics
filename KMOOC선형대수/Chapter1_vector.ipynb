### 일반적인 평면의 방정식을 이용하여 3 차원 simulation

평면의 방정식 ax+by+cz=d 를 이용한다. a, b, c 는 평면의 normal vecor (법선벡터) 에서 구하고, d 는 별도로 계산한다. 법선 vector 는 점들을 연결하는 두 vector 의 외적 (cross-product) 에서 구하고, d 는 normal vector 와 다른 임의의 위치 vector (position vector) 와의 내적(dot product) 으로 구한다.

```python
import numpy as np

p1 = np.array([4, -3, 1])
p2 = np.array([6, -4, 7])
p3 = np.array([1, 2, 2])
```

```python
# These two vectors are in the plane
v1 = p3 - p1
v2 = p2 - p1

# the cross product is a vector normal to the plane
cp = np.cross(v1, v2)
a, b, c = cp

# This evaluates a * x3 + b * y3 + c * z3 which equals d
d = np.dot(cp, p3)

print('The equation is {0}x + {1}y + {2}z = {3}'.format(a, b, c, d))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-2, 14, 10)
y = np.linspace(-2, 14, 10)
X, Y = np.meshgrid(x, y)

Z = (d - a * X - b * Y) / c

# plot the mesh. Each array is 2D, so we flatten them to 1D arrays
ax.plot(X.flatten(),
        Y.flatten(),
        Z.flatten(), 'bo')

# plot the original points. We use zip to get 1D lists of x, y and z
# coordinates.
ax.plot(*zip(p1, p2, p3), color='r', linestyle=' ', marker='o')

# adjust the view so we can see the point/plane alignment
ax.view_init(15, 45)
plt.tight_layout()
#plt.savefig('images/plane.png')
plt.show()
```

### 정사영 (projection)

x = (2, -1, 3), y = (4, -1, 2) 에 대하여 x 위로의 y 의 정사영 $proj_{x}y$ 와 x 에 수직인 y 의 vector 성분 w 를 구하라.

```python
x = vector(2, -1, 3)
y = vector(4, -1, 2)
yx = np.dot(y, x)
xx = np.dot(x, x)

p = yx / xx * x
w = y - p
print(w, 13/7, 1/14, -17/14)
```
