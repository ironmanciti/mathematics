# 3. Multivariate Calculus (다변수 미적분)

### derivative (도함수)의 정의
1. 기하학적 정의 - derivative 는 graph 의 기울기 (slope) 이다.  
    곡선에서 접선(tangent line)의 기울기 즉, 특정 point 에서의 변화 정도.
    
2. 물리학적 정의 - 변화율(rate of change)  
    얼마나 빠르게 어느 방향으로 변하는가. 즉, 가속도.
    
### derivative 의 연산자 : $\frac{d}{dx}$  
If y = f(x), then $$\frac{dy}{dx} = \frac{df(x)}{dx} = \frac{d}{dx}f(x)$$

#### scalar (스칼라, 크기만 있고 방향을 가지지 않는 양)의 derivative 는 항상 0 이다.

### power rule

$$\frac{d}{dx}x^n = nx^{n-1}$$

### multiplication

$$\frac{d}{dx}cx^n = c\frac{d}{dx}x^n = cnx^{n-1}$$

### sum rule
$$\frac{d}{dx}(f(x) + g(x)) = \frac{d}{dx}f(x) + \frac{d}{dx}g(x)$$

### product rule
$$\frac{d}{dx}(f(x)g(x)) = g(x)\frac{d}{dx}f(x) + f(x)\frac{d}{dx}g(x)$$

### chain rule
$$\frac{d}{dx}f(g(x)) = f'(g(x))g'(x)$$

## 편미분(Partial Derivative) - 하나 이상의 변수로 구성된 함수를 미분할 때 사용 
    machine learning 에서는 하나의 변수만 미분하는 경우가 거의 없음
ex) $f(x, y) = 3x^2y$  

$\displaystyle \frac{\partial}{\partial x}f(x, y) = \frac{\partial}{\partial x}3x^2y = 3y\frac{\partial}{\partial x}x^2 = 6xy $  
$\displaystyle \frac{\partial}{\partial y}f(x, y) = \frac{\partial}{\partial y}3x^2y = 3x^2\frac{\partial}{\partial y}y = 3x^2 $

#### anti-derivative (원함수, 역도함수)
$\int \limits_a^b f'(x)dx = f(x) + C$  
(x = b 일 때의 f(x) + C 값) - (x = a 일 때의 f(x) + C 값)

$\int \limits_a^b f'(x)dx = f(b)  - f(a)$


# Gradients (구배, 공간의 기울기)

- partial derivatives 는 vector 로 구성할 수 있다.
$$\frac{d}{dx} \text{ and } \frac{d}{dy}$$
$$\Downarrow$$
$$\nabla = \begin{bmatrix}
            \displaystyle \frac{d}{dx},\frac{d}{dy}
            \end{bmatrix}$$
            
- Gradient of f(x, y)
$$\nabla f(x,y) = \begin{bmatrix}\displaystyle \frac{df(x,y)}{dx}, \frac{df(x,y)}{dy} \end{bmatrix}$$

ex) $\text{ } f(x,y) = 3x^2y $  

$$\nabla 3x^2y = \begin{bmatrix}\displaystyle \frac{3x^2y}{dx}, \frac{3x^2y}{dy} \end{bmatrix} = \begin{bmatrix}6xy, 3x^2 \end{bmatrix}$$

- Gradients 가 derivatives 의 vector 이므로, gradients 는 함수 f(x,y)의 slope 을 나타낸다. 즉, 함수가 가장 가파르게 증가하는 방향을 가리키고, 그 크기(magnitude)는 그 방향으로의 기울기(slope)를 나타낸다.

- Gradient of f(x, y, z)
$$\nabla f(x,y,z) = \begin{bmatrix}\displaystyle \frac{df(x,y,z)}{dx}, \frac{df(x,y,z)}{dy}, \frac{df(x,y,z)}{dz}\end{bmatrix}$$

### Jacobian Matrix (야코비안 행렬 : 다변수 함수의 도함수 행렬)
- multiple function 의 partial derivative matrix (함수가 하나인 경우는 Jacobian == 함수의 gradient)

- n feature 로 이루어진 m sample 의 gradient 는 개개의 sample 을 하나의 함수로 가정하여 Jacobian matrix 로 표시할 수 있다.

$$J = \begin{bmatrix}
        \nabla f(x, y)\\
        \nabla g(x, y)
       \end{bmatrix}  = \begin{bmatrix}
                           \displaystyle \frac{df(x,y)}{dx} \text{  } \frac{df(x,y)}{dy} \\
                           \displaystyle \frac{dg(x,y)}{dx} \text{  } \frac{dg(x,y)}{dy}
                         \end{bmatrix}$$
                         
ex)$\text{  }f(\textbf{x})\text{ where }\textbf{x} = \begin{bmatrix}
                                x1\\
                                x2\\
                                \cdot\\
                                \cdot\\
                                x_n
                              \end{bmatrix}$
                                

$J = \begin{bmatrix}
       \nabla f_1(\textbf{x})\\
       \nabla f_2(\textbf{x})\\
       \text{......}\\
       \nabla f_m(\textbf{x})
     \end{bmatrix} = \begin{bmatrix}
                         \frac{d}{dx_1}f_1(\textbf{x}) \frac{d}{dx_2}f_1(\textbf{x})\cdot \cdot \cdot \frac{d}{dx_n}f_1(\textbf{x})\\
                         \frac{d}{dx_1}f_2(\textbf{x}) \frac{d}{dx_2}f_2(\textbf{x})\cdot \cdot \cdot \frac{d}{dx_n}f_2(\textbf{x})\\
                         \text{......}\\ 
                         \frac{d}{dx_1}f_m(\textbf{x}) \frac{d}{dx_2}f_m(\textbf{x})\cdot \cdot \cdot \frac{d}{dx_n}f_m(\textbf{x})\\
                      \end{bmatrix}
                      = \begin{bmatrix}
                         \frac{\partial}{\partial x_1}f_1 \frac{\partial}{\partial x_2}f_1\cdot \cdot \cdot \frac{\partial}{\partial x_n}f_1\\
                          \frac{\partial}{\partial x_1}f_2 \frac{\partial}{\partial x_2}f_2\cdot \cdot \cdot \frac{\partial}{\partial x_n}f_2\\
                         \text{......}\\ 
                          \frac{\partial}{\partial x_1}f_m \frac{\partial}{\partial x_2}f_m\cdot \cdot \cdot \frac{\partial}{\partial x_n}f_m\\
                      \end{bmatrix}$

# Gradient Visualization

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
print(matplotlib.__version__)
```

```python
nx, ny = (100, 100)
x = np.linspace(0, 10, nx)
y = np.linspace(0, 10, ny)
xv, yv = np.meshgrid(x, y)
```

```python
# define a function
def f(x, y):
    return x * y**2

z = f(xv, yv)
```

```python
plt.figure(figsize=(10,8))
plt.pcolor(xv, yv, z)
plt.colorbar()
plt.title('2D color plot of f(x,y) = x * y^2')
plt.show()
```

```python
x = np.linspace(0, 10, 10)
y = np.linspace(0, 10, 10)
xx, yy = np.meshgrid(x, y)
```

```python
Gy, Gx = np.gradient(f(xx,yy))   # numpy gradient 는 각 dimension 의 axis 순서로 return 하므로 y, x 순서 주의
```

```python
plt.figure(figsize=(10,8))
plt.pcolor(xv, yv, z)
plt.colorbar()
plt.quiver(xx, yy, Gx, Gy, color='w')
plt.title('2D color plot of f(x,y) = x * y^2 with its gradient')
plt.show()
```

```python
# verify numpy gardient by manual calculation of f(x,y) = xy^2
def ddx(x,y):  
    return y**2

def ddy(x,y):
    return 2 * x * y

Gx = ddx(xx, yy)
Gy = ddy(xx, yy)
```

```python
plt.figure(figsize=(10,8))
plt.pcolor(xv, yv, z)
plt.colorbar()
plt.quiver(xx, yy, Gx, Gy, color='w')
plt.title('2D color plot of f(x,y) = x * y^2 with its gradient')
plt.show()
```
